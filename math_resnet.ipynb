```python
from torch.utils.data import Dataset

class MyDataLoader(Dataset):
    def __init__(self, txt, transform=None, target_transform=None, loader=MyLoader):
        imgs = []
        with open(txt) as f:
            for line in f.readlines():
                line = line.strip('\n')  # 移除字符串首尾的换行符
                line = line.rstrip()  # 删除末尾空
                words = line.split(' ')  # 以空格为分隔符 将字符串分成
                imgs.append((words[0], int(words[1]))) # imgs中包含有图像路径和标签
        self.imgs = imgs
        self.transform = transform
        self.target_transform = target_transform
        self.loader = loader
        
    def __getitem__(self, index):
        fn, label = self.imgs[index]
        img = self.loader(fn)
        if self.transform is not None:
            img = self.transform(img)
            
        if self.target_transform is not None:
            label = self.target_transform(label)
        return img.type(torch.float32), label

    def __len__(self):
        return len(self.imgs)
```

```python
from tqdm import tqdm

def image2txt(path, class2idx, r):
    file_list = []
    for file in tqdm(os.listdir(path)):
        label = class2idx[file]
        for img in os.listdir(os.path.join(path, file)):
            file_list.append(os.path.join(path, file, img) + ' ' + str(label) + '\n')
            
    train = open('train.txt', 'w')
    test = open('test.txt', 'w')
    for file in file_list:
        if randint(0, 10) >= r:
            train.write(file)
        else:
            test.write(file)
    train.close()
    test.close()
```

```python
from torch import nn

model.conv1 = nn.Conv2d(1, 64, kernel_size=(7, 7), stride=2, padding=3)
model.fc = nn.Linear(512, 82)
model
```

```python
def accuracy(y_hat, y):
    if len(y_hat.shape) > 1 and y_hat.shape[1] > 1:
        y_hat =  torch.argmax(y_hat, axis=1)
        
    cmp = y_hat.type(y.dtype)==y
    return float(cmp.type(y.dtype).sum())
```

```python
def train(net, train_ter, test_iter, num_epochs, lr, device, threshold):
    def init_weights(m):
        if isinstance(m, nn.Linear) or isinstance(m, nn.Conv2d):
            nn.init.xavier_uniform_(m.weight)
    
    net.apply(init_weights)
    print("device in : ", device)
    net = net.to(device)
    
    loss = nn.CrossEntropyLoss()
    history = {'train_loss':[], 'train_acc':[], 'test_loss':[], 'test_acc':[]}
    for epoch in range(num_epochs):
        net.train()
        train_loss = 0.0
        train_acc = 0.0
        data_num = 0
        optimizer = torch.optim.SGD(net.parameters(), lr=lr * (num_epochs / (num_epochs + epoch / 2)))
        with tqdm(range(len(train_iter)), ncols=100, colour='red', desc="resNet train epoch {}/{}".format(epoch+1, num_epochs)) as pbar:
            for i, (X, y) in enumerate(train_iter):
                optimizer.zero_grad()
                X, y = X.to(device), y.to(device)
                y_hat = net(X)
                l = loss(y_hat, y)
                l.backward()
                optimizer.step()
                train_loss += l.detach()
                train_acc += accuracy(y_hat.detach(), y.detach())
                data_num += X.shape[0]
                pbar.set_postfix({'loss':"{:.2f}".format(l.detach()), 'acc':"{:.2f}".format(train_acc / data_num)})
                pbar.update(1)
                
        history['train_loss'].append(float(train_loss / data_num))
        history['train_acc'].append(float(train_acc / data_num))
  

        net.eval()
        test_loss = 0.0
        test_acc = 0.0
        data_num = 0
        with tqdm(range(len(test_iter)), ncols=100, colour='blue', desc="resNet test epoch {}/{}".format(epoch+1, num_epochs)) as pbar:
            for X, y in test_iter:
                X, y = X.to(device), y.to(device)
                y_hat = net(X)
                with torch.no_grad():
                    l = loss(y_hat, y)
                    test_loss += l.detach()
                    test_acc += accuracy(y_hat.detach(), y.detach())
                    
                    data_num += X.shape[0]
                    pbar.set_postfix({'loss':"{:.2f}".format(l.detach()), 'acc':"{:.2f}".format(test_acc / data_num)})
                    pbar.update(1)
                    
        history['test_loss'].append(float(test_loss / data_num))
        history['test_acc'].append(float(test_acc / data_num))
        if history['test_acc'][-1] > threshold:
            print("early stop")
            break
        
    return history
```

```python
threshold = 0.98
history = train(model, train_iter, test_iter, 20, 5e-3, 'cuda', threshold)
```

```python
import matplotlib.pyplot as plt

train_loss = history['train_loss']
train_acc = history['train_acc']
test_loss = history['test_loss']
test_acc = history['test_acc']
plt.plot(range(len(train_loss)), train_loss, label='train loss')
plt.plot(range(len(train_acc)), train_acc, label='train acc')
plt.plot(range(len(test_loss)), test_loss, label='test loss')
plt.plot(range(len(test_acc)), test_acc, label='test acc')
plt.xlabel('epochs')
plt.ylabel('scores')
plt.legend()
plt.title("resNet")
plt.show()
```

```python
from random import sample
def testModel(net, txt, num_choice):
    with open(txt) as f:
        datas = f.readlines()
    
    data_samples = sample(datas, num_choice)
    images = torch.tensor([])
    predicts = []
    labels = []
    for data in data_samples:
        data = data.strip('\n')
        data = data.rstrip()
        X_path, label = data.split(' ')
        labels.append(label)
        image = io.read_image(X_path)
        image = torch.unsqueeze(image, dim=0)
        images = torch.cat([images, image.type(torch.float32)], dim=0)
  
    predicts = torch.argmax(net(torch.tensor(images).to('cuda')), dim=1)
        
    return images, labels, predicts 
```
